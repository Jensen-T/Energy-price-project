```python
df = pd.read_csv("energy_dataset.csv",parse_dates=['time'],index_col='time')

df.index = pd.to_datetime(df.index, utc = True)

df.columns = df.columns.str.replace(' ','_').str.replace('-','_')

to_drop = ['generation_fossil_coal_derived_gas',
           'generation_fossil_oil_shale',
           'generation_fossil_peat',
           'generation_geothermal',
           'generation_hydro_pumped_storage_aggregated',
           'generation_marine',
           'generation_wind_offshore',
           'forecast_wind_offshore_eday_ahead'] #these are all columns that had majority 0 or NaN values

df = df.drop(columns = to_drop)
```

```python
df.isnull().sum().sort_values(ascending = False)
```

<li> There are not a large number NaN values compared to the data size of 35064, so they will not severely affect the dataset, and thus can be removed </li>

```python
df = df.dropna() #we will thus drop the rows with NaN values in them </li>
```

```python
df.columns
```

```python
x = df[['generation_biomass', 'generation_fossil_brown_coal/lignite',
       'generation_fossil_gas', 'generation_fossil_hard_coal',
       'generation_fossil_oil', 'generation_hydro_pumped_storage_consumption',
       'generation_hydro_run_of_river_and_poundage',
       'generation_hydro_water_reservoir', 'generation_nuclear',
       'generation_other', 'generation_other_renewable', 'generation_solar',
       'generation_waste', 'generation_wind_onshore',
       'forecast_solar_day_ahead', 'forecast_wind_onshore_day_ahead',
       'total_load_forecast', 'total_load_actual', 'price_day_ahead']]

y = df[['price_actual']] #setting up independent vs dependent variables
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() #first we will trial using multiple regression
model.fit(x_train, y_train)

y_prediction = model.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

score = r2_score(y_test, y_prediction)

print('r2 score is', score)
print('mean_sqrd_error is =', mean_squared_error(y_test, y_prediction))
print('root_mean_squared error of is =', np.sqrt(mean_squared_error(y_test,  y_prediction)))
```

<li> We can see that the r2 score is very low, which means that we shouldn't use this to predict, and it is quite far below 0.8 </li>
